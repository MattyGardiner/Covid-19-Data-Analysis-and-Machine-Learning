# There is a subtle inflexion point from 'concave upward' to 'concave downward' suggesting the relationship between 'TotalPositiveCases' and 'Deaths' is first accelerating then decelerating. The point of inflection would be of particular interest. It would be interesting to see what data coincides with this point of inflection, e.g. total % of Italian population that have tested positive. Usually, the point of inflection occurs where total % of population which have tested positive for a virus exceeds 50%. To see where the point of inflection occurs we must, using calculus, see where the second derivative changes sign. Typically increased immune activation proceeds an point of inflection. 




```python
sns.relplot(y="Recovered", x="TotalPositiveCases", data=data_region)
```

```python
sns.relplot(y="HospitalizedPatients", x="TotalPositiveCases",hue="Recovered", data=data_region)
```
